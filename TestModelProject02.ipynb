# **: Clean & Change Type Data**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('DataTrain_Model63.xlsx', sheet_name='datacleantoUSE')
df
```

```python
df.info()
```

```python
#Change Data Type and Replace Value

df['Sex'] = df['Sex'].replace({'M': 1, 'F': 0})
df['Smoke'] = df['Smoke'].replace({'Yes': 1, 'No': 0})
df['Drink'] = df['Drink'].replace({'Yes': 1, 'No': 0})
df.head()
```

```python
def create_age_range(df):               #Create Age Range: create new column Age Range
  conditions = [
        (df['Age'] < 20),
        (df['Age'] >= 20) & (df['Age'] <= 30),
        (df['Age'] >= 31) & (df['Age'] <= 40),
        (df['Age'] >= 41) & (df['Age'] <= 50),
        (df['Age'] >= 51)]
  values = [0, 1, 2, 3, 4]
  df['Age_Range'] = pd.Series(pd.cut(df['Age'], bins=[-float('inf'), 20, 31, 41, 51, float('inf')], labels=values, right=False))
  return df

df = create_age_range(df)
df['Age_Range']
```

```python
def create_BMI_range(df):               #Create BMI Range: create new column BMI Range
  conditions = [
        (df['BMI'] < 18.50),
        (df['BMI'] >= 18.50),
        (df['BMI'] >= 23.01),
        (df['BMI'] >= 25.00),
        (df['BMI'] >= 30.00)]
  values = [1, 2, 3, 4, 5]
  df['BMI_Range'] = pd.Series(pd.cut(df['BMI'], bins=[-float('inf'), 18.50, 23.01, 25.00, 30.00, float('inf')], labels=values, right=False))
  return df

df = create_BMI_range(df)
df['BMI_Range']
```

```python
def create_CDISESE_range(df):               #Create CDISESE Range: create new column CDISESE Range
  conditions_cdisese = [
        (df['CDISESE'] < 1),
        (df['CDISESE'] >= 1),
        (df['CDISESE'] >= 2),
        (df['CDISESE'] >= 3),
        (df['CDISESE'] >= 4),
        (df['CDISESE'] > 50)]
  values = [0, 3, 1, 2, 5, 4]
  df['CDISESE_Range'] = pd.Series(pd.cut(df['CDISESE'], bins=[-float('inf'), 1, 2, 3, 4, 50, float('inf')], labels=values, right=False))
  return df

df = create_CDISESE_range(df)
df['CDISESE_Range'].head(30)
```

```python
#Select Column: create DataFrame to Moedling
columns_to_select = ['Sex',	'BP',	'Height',	'Weight',	'Smoke', 'Drink',	'TRI_TGBLOOD',	'FBS_BLOOD',	'SGOT',	'SGPT',
                     'ALP',	'BUN',	'CREATININE',	'URIC_ACID',	'UA',	'CBC',	'Age_Range', 'BMI_Range', 'CDISESE_Range',	'DLD_CLASS']
df2 = df[columns_to_select].copy()
df2
```

```python
df2.info()
```

```python
df2['Age_Range'] = df2['Age_Range'].astype(int)
df2['BMI_Range'] = df['BMI_Range'].astype(int)
df2['CDISESE_Range'] = df2['CDISESE_Range'].astype(int)
df2.info()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
import lightgbm as lgb
```

```python
x = df2.drop('DLD_CLASS', axis=1)   # Features
y = df2['DLD_CLASS']                # Target variable

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

# Initialize classifiers
classifiers = {
    'Random Forest': RandomForestClassifier(random_state = 42),
    'Decision Tree': DecisionTreeClassifier(random_state = 42),
    'Logistic Regression': LogisticRegression(random_state = 42),
    'XGBoost': XGBClassifier(random_state = 42),
    'LightGBM': lgb.LGBMClassifier(random_state = 42)
    }
```

```python
x_train.info()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score

# Train and evaluate each classifier
for name, clf in classifiers.items():

    # Train the model
    clf.fit(x_train, y_train)

    # Make predictions on the test set
    y_pred = clf.predict(x_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    classification_report_result = classification_report(y_test, y_pred)
    confusion = confusion_matrix(y_test, y_pred)  # Change y_true to y_test

    # Display the evaluation metrics
    print(f"\nEvaluation for {name}:")
    print(f"Accuracy: {accuracy}")
    print("Classification Report:")
    print(classification_report_result)
    print(confusion)

    # Cross-validation scores
    cv_scores = cross_val_score(clf, x, y, cv=5)
    print(f"Cross-validation scores for {name}: {cv_scores}")
    print(f"Mean Cross-validation score for {name}: {cv_scores.mean()}")
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

# Create Model CatBoostClassifier
cat = CatBoostClassifier(random_state=42)

# Train the model
cat.fit(x_train, y_train)

# Make predictions on the test set
y_pred = cat.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_result = classification_report(y_test, y_pred)
confusion01 = confusion_matrix(y_test, y_pred)


# Display the evaluation metrics
print(f"\nEvaluation for CatBoost:")
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report_result)
print(confusion01)
```

```python
# Cross-validation scores for CatBoost
cv_scores = cross_val_score(cat, x, y, cv=5)
print(f"Cross-validation scores for CatBoost: {cv_scores}")
print(f"Mean Cross-validation score for CatBoost: {cv_scores.mean()}")
```
